## What this notebook does

The previous notebook (`Jaggaer_Data_EDA_v2.ipynb`) proved the retrieval concept on a single product category (serological pipettes). This notebook operationalizes that into a working end-to-end system.

### Architecture

```
User query (natural language)
        |
        v
[TIER 1: Exact match]  (mfr + part number)
   - Deterministic, zero LLM cost
   - 42,052 cross-distributor matches already known
   - If match found: return best price + alternatives immediately
        |
        v  (if no exact match)
[TIER 2: Semantic retrieval]  (FAISS cosine, threshold 0.85)
   - Full catalog FAISS index (bge-small-en-v1.5, 384-dim)
   - emb_text = category_desc | manufacturer | description
   - Returns top-K candidates (median ~44, max ~137 at thresh 0.85)
        |
        v
[TIER 3: LLM reasoning]  (qwen2.5:7b or nemotron-3-nano)
   - Receives candidate table (distributor, mfr, description, price)
   - Task: identify equivalents, pick cheapest, output structured JSON
   - Two endpoints tested side-by-side (Ollama local vs NIM server)
        |
        v
Structured answer: best price, distributor, alternatives, savings vs worst
```

### Questions this notebook answers for stakeholders

1. **Can we build a RAG?** Yes - demonstrated end to end below.
2. **Small normalizing model or large reasoning model?** Benchmarked both.
3. **How fast is each path?** Measured: exact-match (< 50ms), semantic (< 2s), LLM (2-12s).
4. **How much can we save without AI at all?** Deterministic part-number matching already proves &#36;317K+ floor.
5. **Is this scalable?** FAISS at 6M rows fits in ~1GB RAM (quantized). LLM cost = 0 on institutional NIM.

## Stage 11: Full Catalog FAISS Index

The Stage 10 pilot embedded only serological pipettes (~few hundred rows). Here we embed the **entire catalog** (6M rows, deduped to unique descriptions).

### Why bge-small-en-v1.5
- 384-dim vectors (vs 768 for bge-large): index size ~9GB float32 -> ~900MB IVF-quantized
- Proven in Stage 10 to produce clean separations on lab supply descriptions
- Fast: ~100K rows/min on GPU

### Runtime estimate
- 6M rows, ~40% dedup rate -> ~3.6M unique texts to embed
- At 100K/min on GPU: ~36 minutes
- **This cell saves the index to disk so you only run it once.** Subsequent cells load from disk.

### Index format
We use `faiss.IndexIVFFlat` with `nlist=4096` cells and inner-product (cosine on normalized vectors). This gives sub-second search on 3.6M vectors vs the full exhaustive scan of `IndexFlatIP`.

## Stage 12: Retrieval Function

Two-tier retrieval. Tier 1 tries deterministic part-number matching first (instant, no AI cost). Tier 2 falls back to semantic search if nothing is found.

### Part-number match table
Pre-built from Stage 8: 42,052 `(mfr_key, part_key)` pairs that appear in 2+ distributors. When a query hits this table, we already know the cheapest option without any embedding or LLM work.

### Semantic retrieval
- Embed the query with `bge-small-en-v1.5` (BGE query prefix)
- Range-search FAISS at cosine threshold 0.85
- Join results back to `df_all` for full row metadata
- Cap at 150 candidates before handing to LLM

````python
# 12.1 - Build the exact-match lookup table from df_all
# This replicates the Stage 8 logic but produces a reusable dict.

def norm_key(s):
    if pd.isna(s): return None
    return str(s).strip().upper()

with tee_output('48_exact_match_table'):
    print('=== Stage 12.1: Exact-Match Lookup Table ===')
    print()
    print('Building exact-match lookup table...')
    t_start = time.time()

    keyed = df_all[
        df_all['manufacturer'].notna() &
        df_all['mfr_part_num'].notna() &
        (df_all['price_num'] > 0) &
        (df_all['price_num'] < 10_000_000)
    ].copy()
    keyed['mfr_key']  = keyed['manufacturer'].apply(norm_key)
    keyed['part_key'] = keyed['mfr_part_num'].apply(norm_key)
    keyed = keyed[keyed['mfr_key'].notna() & keyed['part_key'].notna()]
    print(f'Rows with usable (mfr, part, price): {len(keyed):,}')
    print(f'By distributor:')
    print(keyed['distributor'].value_counts().to_string())

    # One row per (mfr, part, distributor) - cheapest price each
    exact_table = (
        keyed
        .groupby(['mfr_key', 'part_key', 'distributor'], as_index=False)
        .agg(price=('price_num', 'min'),
             description=('description', 'first'),
             manufacturer=('manufacturer', 'first'),
             dist_catalog_num=('dist_catalog_num', 'first'),
             qty=('qty_num', 'max'))
    )
    print(f'  groupby done in {time.time()-t_start:.1f}s')

    # Which (mfr, part) appear in 2+ distributors?
    span = exact_table.groupby(['mfr_key', 'part_key'])['distributor'].nunique()
    multi_keys = set(span[span >= 2].index.tolist())
    print(f'\nTotal (mfr, part, distributor) rows: {len(exact_table):,}')
    print(f'(mfr, part) keys in 2+ distributors: {len(multi_keys):,}')
    print(f'(mfr, part) keys in only 1 distrib:  {len(span) - len(multi_keys):,}')

    # Build O(1) dict - fast version using itertuples (no Python groupby loop)
    print('Building exact_dict...')
    t_dict = time.time()
    sorted_table = exact_table.sort_values(['mfr_key', 'part_key', 'price'])
    exact_dict = {}
    cols = ['mfr_key','part_key','distributor','price','description','manufacturer','dist_catalog_num','qty']
    for row in sorted_table[cols].itertuples(index=False):
        key = (row.mfr_key, row.part_key)
        rec = {
            'mfr_key': row.mfr_key, 'part_key': row.part_key,
            'distributor': row.distributor, 'price': row.price,
            'description': row.description, 'manufacturer': row.manufacturer,
            'dist_catalog_num': row.dist_catalog_num, 'qty': row.qty,
        }
        if key not in exact_dict:
            exact_dict[key] = []
        exact_dict[key].append(rec)
    print(f'  exact_dict: {len(exact_dict):,} keys in {time.time()-t_dict:.1f}s')

    # Spot-check: show 5 sample cross-distributor matches
    sample_keys = list(multi_keys)[:5]
    print('\nSample cross-distributor matches:')
    for mk, pk in sample_keys:
        rows = exact_dict.get((mk, pk), [])
        print(f'  {mk[:25]} | {pk[:20]}')
        for r in rows:
            print(f'    [{r["distributor"]:5s}] ${r["price"]:>10.2f}  {str(r["description"])[:55]}')

    print(f'\nExact-match table ready. Total time: {time.time()-t_start:.1f}s')
````

````python
# 12.2 - Retrieval functions

def search_exact(mfr: str = None, part_num: str = None) -> pd.DataFrame:
    """
    Exact match on (manufacturer, part_number).
    Uses pre-built dict for O(1) lookup - should return in <1ms.
    Returns a dataframe sorted cheapest first, or empty if no match.
    """
    if not mfr or not part_num:
        return pd.DataFrame()
    mk = norm_key(mfr)
    pk = norm_key(part_num)
    rows = exact_dict.get((mk, pk), [])
    if not rows:
        return pd.DataFrame()
    return pd.DataFrame(rows)  # already sorted by price at build time


def search_semantic(query: str,
                    threshold: float = COSINE_THRESHOLD,
                    max_candidates: int = 150) -> pd.DataFrame:
    """
    Semantic retrieval via FAISS range search.
    Returns up to max_candidates rows sorted by descending cosine similarity.
    Joins back to df_all for full metadata.
    """
    t0 = time.time()
    qv = embed_query(query)

    # range_search returns all vectors with similarity >= threshold
    lims, sims_flat, ids_flat = index.range_search(qv, threshold)
    n_hits = int(lims[1] - lims[0])

    if n_hits == 0:
        return pd.DataFrame()

    sims = sims_flat[lims[0]:lims[1]]
    ids  = ids_flat[lims[0]:lims[1]]

    # sort by similarity descending, cap at max_candidates
    order = np.argsort(-sims)
    sims  = sims[order][:max_candidates]
    ids   = ids[order][:max_candidates]

    result = meta.iloc[ids].copy()
    result.insert(0, 'cosine', sims)
    result = result.reset_index(drop=True)

    elapsed = time.time() - t0
    print(f'  semantic: {n_hits} hits at cos>={threshold:.2f}, '
          f'returning top {len(result)}, took {elapsed*1000:.0f}ms')
    return result


def retrieve(query: str,
             mfr: str = None,
             part_num: str = None) -> dict:
    """
    Two-tier retrieval entry point.
    Returns dict with keys:
      'tier'         : 'exact' or 'semantic'
      'candidates'   : DataFrame of candidate products
      'query'        : the original query string
    """
    # --- Tier 1: exact match ---
    if mfr and part_num:
        exact = search_exact(mfr, part_num)
        if len(exact) >= 2:   # 2+ distributors carry it -> we have a comparison
            return {'tier': 'exact', 'candidates': exact, 'query': query}

    # --- Tier 2: semantic ---
    semantic = search_semantic(query)
    return {'tier': 'semantic', 'candidates': semantic, 'query': query}


print('Retrieval functions defined.')
print('  search_exact(mfr, part_num)  - deterministic lookup')
print('  search_semantic(query)       - FAISS range search')
print('  retrieve(query, mfr, part)   - two-tier entry point')
````

## Stage 13: LLM Reasoning Layer - Ollama vs NIM

The retrieval layer returns a candidate table. The LLM's job is to:
1. Identify which candidates are equivalent to what the user asked for
2. Return the cheapest option per distributor
3. Estimate savings vs the most expensive equivalent
4. Output structured JSON

We test **two endpoints side by side**:
- **Ollama / qwen2.5:7b** (local): won the model benchmark (file 44) - 4/4 correct, no wrong merges, 3.1s avg
- **NIM / nemotron-3-nano** (server): existing institutional infra, 256K context window

The benchmark records: parse success rate, answer correctness, and wall-clock latency per endpoint.

````python
# 13.1 - LLM client with caching (same pattern as EDA notebook)

LLM_CACHE_PATH = Path('analysis_output/llm_cache_rag.json')

def _prompt_hash(system_prompt, user_prompt):
    combined = f'{system_prompt}|||{user_prompt}'
    return hashlib.sha256(combined.encode('utf-8')).hexdigest()[:16]

def _load_cache():
    if LLM_CACHE_PATH.exists():
        try: return json.load(open(LLM_CACHE_PATH, encoding='utf-8'))
        except: return {}
    return {}

def _save_cache(cache):
    json.dump(cache, open(LLM_CACHE_PATH, 'w', encoding='utf-8'), indent=2)


def call_llm(base_url: str, model: str,
             system_prompt: str, user_prompt: str,
             max_tokens: int = 800,
             temperature: float = 0.1,
             use_cache: bool = True) -> dict:
    """
    Call an OpenAI-compatible LLM endpoint.
    Returns dict with keys: content, elapsed_sec, cached, error.
    """
    cache = _load_cache()
    key = _prompt_hash(system_prompt, user_prompt)

    if use_cache and key in cache:
        return {'content': cache[key], 'elapsed_sec': 0.0, 'cached': True, 'error': None}

    payload = {
        'model': model,
        'messages': [
            {'role': 'system', 'content': system_prompt},
            {'role': 'user',   'content': user_prompt},
        ],
        'max_tokens': max_tokens,
        'temperature': temperature,
    }
    url = base_url.rstrip('/') + '/chat/completions'
    t0 = time.time()
    try:
        resp = requests.post(url, json=payload, timeout=90)
        resp.raise_for_status()
        content = resp.json()['choices'][0]['message']['content']
        elapsed = time.time() - t0
        if content and content.strip():
            cache[key] = content
            _save_cache(cache)
        return {'content': content, 'elapsed_sec': elapsed, 'cached': False, 'error': None}
    except Exception as e:
        return {'content': None, 'elapsed_sec': time.time()-t0, 'cached': False, 'error': str(e)}


def check_endpoint(base_url, model, name):
    """Quick connectivity check."""
    result = call_llm(base_url, model,
                      'You are a test.', 'Reply: OK',
                      max_tokens=10, use_cache=False)
    if result['error']:
        print(f'  {name}: OFFLINE - {result["error"][:80]}')
    else:
        print(f'  {name}: ONLINE  - replied in {result["elapsed_sec"]:.2f}s')
    return result['error'] is None


print('Checking LLM endpoints...')
ollama_online = check_endpoint(OLLAMA_URL, OLLAMA_MODEL, 'Ollama/qwen2.5:7b')
nim_online    = check_endpoint(NIM_URL,    NIM_MODEL,    'NIM/nemotron-3-nano')
````

````python
# 13.2 - Reasoning prompt + JSON parser

SYSTEM_PROMPT = """\
You are a procurement assistant for a university research lab.
You receive a list of candidate products from a catalog database.
Your job:
1. Identify which candidates are equivalent (same physical product or interchangeable).
2. Find the cheapest price across all equivalent products.
   The 'best' field MUST be the single lowest-priced equivalent product.
   Sort alternatives by price ascending. Never put a higher-priced item in 'best'.
3. Return your answer as JSON only - no preamble, no explanation, just the JSON object.

Output schema (strict):
{
  "best": {
    "distributor": str,
    "manufacturer": str,
    "description": str,
    "price": float,
    "catalog_num": str
  },
  "alternatives": [
    {"distributor": str, "price": float, "description": str}
  ],
  "savings_vs_worst": float,
  "n_equivalent": int,
  "reasoning": str
}

Equivalence rules - a candidate is ONLY equivalent if ALL of these are true:
- PRODUCT TYPE matches the query exactly. If the query asks for a vial, return vials only.
  A cap, lid, adapter, or accessory is NOT equivalent to the product it fits.
  A pipette tip is NOT equivalent to a pipette. A flask is NOT equivalent to a plate.
- VOLUME / SIZE matches. A 5 mL pipette is not equivalent to a 10 mL pipette.
  A 1000 uL tip is not equivalent to a 200 uL tip.
- PACK SIZE: do NOT list different pack sizes as alternatives to each other.
  If comparing prices across pack sizes, normalize to per-unit price first.
  Only compare products at the same pack size as alternatives.
- MATERIAL matches when specified. Polystyrene and glass are not interchangeable.

If fewer than 2 genuinely equivalent products are found after applying these rules,
set alternatives to [] and savings_vs_worst to 0.
Price values must be numeric (no dollar signs).
"""


def format_candidates(candidates: pd.DataFrame, query: str, max_rows: int = 100) -> str:
    """
    Format the candidate DataFrame into a compact string for the LLM prompt.
    Limits to max_rows to stay within context budget.
    """
    rows = candidates.head(max_rows)
    lines = [f'USER QUERY: {query}\n', 'CANDIDATE PRODUCTS:']
    for i, (_, r) in enumerate(rows.iterrows(), 1):
        price = r.get('price_num', r.get('price', 'N/A'))
        cat_num = r.get('dist_catalog_num', r.get('catalog_num', ''))
        lines.append(
            f'{i:3d}. [{r["distributor"]:5s}] '
            f'${price:<10.2f} '
            f'MFR:{str(r["manufacturer"])[:30]:<32} '
            f'CAT:{str(cat_num)[:20]:<22} '
            f'{str(r["description"])[:90]}'
        )
    return '\n'.join(lines)


def parse_llm_json(content: str) -> dict | None:
    """
    Extract JSON from LLM response. Handles markdown code fences.
    Returns None if parsing fails.
    """
    if not content:
        return None
    # strip markdown fences
    clean = content.strip()
    if '```' in clean:
        parts = clean.split('```')
        for p in parts:
            p = p.strip()
            if p.startswith('json'):
                p = p[4:].strip()
            if p.startswith('{'):
                clean = p
                break
    # find first { ... } block
    start = clean.find('{')
    end   = clean.rfind('}')
    if start == -1 or end == -1:
        return None
    try:
        return json.loads(clean[start:end+1])
    except json.JSONDecodeError:
        return None


print('Prompt and parser defined.')
````

## Stage 14: End-to-End Query Interface

The full pipeline in one callable function: `ask_procurement(query)`.  
This is the thing that a user would call from a web UI or a Jupyter cell.

Outputs a structured result dict and a human-readable summary.

````python
# 14.0 - Helper: guarantee best = cheapest
# Must run before 14.1

def _sort_by_price(best, alternatives):
    """Sort all equivalent products by price, return (cheapest, rest, savings)."""
    all_items = ([best] if best else []) + (alternatives or [])
    all_items = [p for p in all_items if p and p.get('price') is not None]
    all_items.sort(key=lambda p: float(p.get('price', 9e9)))
    if not all_items:
        return None, [], 0
    cheapest = all_items[0]
    rest = all_items[1:]
    savings = round(
        float(all_items[-1].get('price', 0)) - float(cheapest.get('price', 0)), 2
    ) if rest else 0
    return cheapest, rest, savings

print('_sort_by_price defined.')
````

````python
# 14.1 - Full pipeline function

# Which endpoint to use for production queries
# Priority: Ollama (local, fast, proven) > NIM (server, larger context)
ACTIVE_ENDPOINT = None
if ollama_online:
    ACTIVE_ENDPOINT = (OLLAMA_URL, OLLAMA_MODEL, 'Ollama/qwen2.5:7b')
elif nim_online:
    ACTIVE_ENDPOINT = (NIM_URL, NIM_MODEL, 'NIM/nemotron-3-nano')

if ACTIVE_ENDPOINT:
    print(f'Active endpoint: {ACTIVE_ENDPOINT[2]}')
else:
    print('WARNING: No LLM endpoint online. Pipeline will return retrieval results only.')


def ask_procurement(query: str,
                    mfr: str = None,
                    part_num: str = None,
                    verbose: bool = True) -> dict:
    """
    Full procurement query pipeline.

    Parameters
    ----------
    query    : natural language product description
    mfr      : manufacturer name (optional, enables exact-match fast path)
    part_num : manufacturer part number (optional, enables exact-match fast path)
    verbose  : print formatted output

    Returns
    -------
    dict with keys:
        tier, candidates_n, llm_answer, elapsed_retrieval_sec,
        elapsed_llm_sec, best, alternatives, savings_vs_worst
    """
    timings = {}

    # --- Retrieval ---
    t0 = time.time()
    r = retrieve(query, mfr=mfr, part_num=part_num)
    timings['retrieval'] = time.time() - t0
    cands = r['candidates']

    if verbose:
        print(f'Query:      {query}')
        print(f'Tier:       {r["tier"]}  |  candidates: {len(cands)}  |  '
              f'retrieval: {timings["retrieval"]*1000:.0f}ms')

    if len(cands) == 0:
        if verbose: print('No candidates found.')
        return {'tier': r['tier'], 'candidates_n': 0, 'llm_answer': None,
                'elapsed_retrieval_sec': timings['retrieval'],
                'elapsed_llm_sec': 0, 'best': None, 'alternatives': [], 'savings_vs_worst': 0}

    # --- Exact-match short circuit (no LLM needed) ---
    if r['tier'] == 'exact':
        best_row = cands.iloc[0]
        worst_price = cands['price'].max()
        savings = worst_price - best_row['price']
        result = {
            'tier': 'exact',
            'candidates_n': len(cands),
            'llm_answer': None,
            'elapsed_retrieval_sec': timings['retrieval'],
            'elapsed_llm_sec': 0,
            'best': {
                'distributor':  best_row['distributor'],
                'manufacturer': best_row['manufacturer'],
                'description':  best_row['description'],
                'price':        best_row['price'],
                'catalog_num':  best_row.get('dist_catalog_num', ''),
            },
            'alternatives': cands.iloc[1:4][['distributor','price','description']].to_dict('records'),
            'savings_vs_worst': savings,
        }
        if verbose:
            _print_answer(result, query)
        return result

    # --- LLM reasoning (semantic tier) ---
    if ACTIVE_ENDPOINT is None:
        if verbose: print('No LLM available - returning raw candidates only.')
        return {'tier': r['tier'], 'candidates_n': len(cands),
                'llm_answer': None, 'elapsed_retrieval_sec': timings['retrieval'],
                'elapsed_llm_sec': 0, 'best': None, 'alternatives': [], 'savings_vs_worst': 0}

    ep_url, ep_model, ep_name = ACTIVE_ENDPOINT
    user_prompt = format_candidates(cands, query)
    t0 = time.time()
    llm_result = call_llm(ep_url, ep_model, SYSTEM_PROMPT, user_prompt,
                           max_tokens=800)
    timings['llm'] = time.time() - t0
    parsed = parse_llm_json(llm_result['content'])

    # Apply price sort immediately after parse, before anything else
    # This fixes cached results too - cache may have old wrong ordering
    if parsed is not None:
        _best, _alts, _sav = _sort_by_price(
            parsed.get('best'), parsed.get('alternatives', []))
        if _best:
            parsed['best'] = _best
            parsed['alternatives'] = _alts
            parsed['savings_vs_worst'] = _sav

    if parsed is None:
        if verbose:
            print(f'LLM parse failed. Raw: {str(llm_result["content"])[:200]}')
        return {'tier': r['tier'], 'candidates_n': len(cands),
                'llm_answer': llm_result['content'],
                'elapsed_retrieval_sec': timings['retrieval'],
                'elapsed_llm_sec': timings['llm'],
                'best': None, 'alternatives': [], 'savings_vs_worst': 0}

    result = {
        'tier': r['tier'],
        'candidates_n': len(cands),
        'llm_answer': parsed,
        'elapsed_retrieval_sec': timings['retrieval'],
        'elapsed_llm_sec': timings.get('llm', 0),
        'best': parsed.get('best'),
        'alternatives': parsed.get('alternatives', []),
        'savings_vs_worst': parsed.get('savings_vs_worst', 0),
    }
    if verbose:
        _print_answer(result, query)
    return result


def _print_answer(result: dict, query: str):
    """Human-readable output for a procurement query result."""
    print()
    best = result.get('best')
    if best:
        print(f'  BEST PRICE:   ${best.get("price","?"):>10.2f}  from [{best.get("distributor","?")}]')
        print(f'  Manufacturer: {best.get("manufacturer","?")}')
        print(f'  Product:      {str(best.get("description",""))[:90]}')
        if best.get('catalog_num'):
            print(f'  Catalog #:    {best.get("catalog_num")}')
        savings = result.get('savings_vs_worst', 0)
        if savings > 0:
            print(f'  Potential savings vs highest price: ${savings:.2f}')
        alts = result.get('alternatives', [])
        if alts:
            print(f'  Alternatives:')
            for a in alts[:3]:
                print(f'    [{a.get("distributor","?")}] ${a.get("price",0):.2f}  {str(a.get("description",""))[:60]}')
    else:
        print('  No equivalent products identified.')
    total_sec = result['elapsed_retrieval_sec'] + result['elapsed_llm_sec']
    print(f'  Latency: retrieval={result["elapsed_retrieval_sec"]*1000:.0f}ms  '
          f'llm={result["elapsed_llm_sec"]*1000:.0f}ms  '
          f'total={total_sec:.2f}s')


print('ask_procurement() ready.')
````

## Stage 15: ROI Visualizations

Three charts built from the deterministic part-number match data (Stage 8 / file 40):

1. **Savings opportunity distribution** - histogram of price spread % across 42K matched products
2. **Top 20 items by estimated savings** - horizontal bar chart
3. **Spread % by distributor pair** - which vendor pairings have the most gap

These are the charts to show stakeholders. The numbers are defensible because they come from exact part-number matches - no AI, no assumptions.

````python
# 15.1 - Rebuild the value-sizing table (same logic as Stage 8)
import matplotlib
matplotlib.use('Agg')   # non-interactive backend; works in Jupyter and headless
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Color palette from project conventions
C_BLUE   = '#1F4E79'
C_BODY   = '#595959'
C_GREEN  = '#548235'
C_RED    = '#C00000'
C_LBLUE  = '#2E75B6'

with tee_output('49_roi_value_sizing'):
    print('=== Stage 15: ROI Value Sizing ===')
    print()
    print('Rebuilding value-sizing table for visualization...')

    keyed_viz = df_all[
        df_all['manufacturer'].notna() &
        df_all['mfr_part_num'].notna() &
        (df_all['price_num'] > 0) &
        (df_all['price_num'] < 10_000_000)
    ].copy()
    keyed_viz['mfr_key']  = keyed_viz['manufacturer'].apply(norm_key)
    keyed_viz['part_key'] = keyed_viz['mfr_part_num'].apply(norm_key)

    grp = (
        keyed_viz[keyed_viz['mfr_key'].notna() & keyed_viz['part_key'].notna()]
        .groupby(['mfr_key', 'part_key', 'distributor'], as_index=False)
        .agg(price=('price_num', 'min'),
             qty=('qty_num', 'max'),
             desc=('description', 'first'),
             manufacturer=('manufacturer', 'first'))
    )
    span = grp.groupby(['mfr_key','part_key'])['distributor'].nunique()
    multi = span[span >= 2].index
    m = grp.set_index(['mfr_key','part_key']).loc[multi].reset_index()

    agg = (
        m.groupby(['mfr_key','part_key'])
         .agg(n_dist=('distributor','nunique'),
              min_price=('price','min'),
              max_price=('price','max'),
              max_qty=('qty','max'),
              desc=('desc','first'),
              manufacturer=('manufacturer','first'))
         .reset_index()
    )
    agg['spread']     = agg['max_price'] - agg['min_price']
    agg['spread_pct'] = 100 * agg['spread'] / agg['max_price']
    agg['est_savings']= agg['spread'] * agg['max_qty'].fillna(0)

    print(f'Cross-distributor matched keys:  {len(agg):,}')
    print(f'Keys with spread > 0:            {(agg["spread"]>0).sum():,}')
    print(f'Total est. savings (qty-matched): ${agg["est_savings"].sum():,.0f}')
    print(f'Keys with no qty data:           {agg["max_qty"].isna().sum():,}')
    print()
    print('Spread % distribution:')
    print(agg[agg['spread_pct']>0]['spread_pct']
          .describe(percentiles=[.25,.5,.75,.9,.99]).to_string())
````

## Stage 17: Conclusions, Open Questions, and Stakeholder Q&A

---

### What this notebook demonstrates

1. **RAG works end-to-end.** Natural language queries return ranked candidate products with prices from across all four BTAA distributors. The system runs on existing infrastructure with zero API cost.

2. **Exact part-number matching is the highest-ROI, lowest-risk tier.** 42,052 matched products, median spread 13%, proven floor savings of &#36;317K+ on quantity-matched items. This requires no AI and can be deployed as a deterministic tool immediately.

3. **Semantic retrieval extends coverage to the un-matchable catalog.** Vector similarity at cosine threshold 0.85 returns a median of 44 candidates per query (max 137), well within the LLM context window. Retrieval quality on the pilot category (serological pipettes) was clean - all 8 top results were genuinely relevant.

4. **qwen2.5:7b is the right model for the reasoning step.** It is the only model to pass the safety test (no wrong merges) with perfect consistency (4/4) in the benchmark. It runs locally via Ollama with no server dependency.

5. **The original &#36;28,864 three-category savings estimate is a floor, not a ceiling.** Full four-distributor exact matching identifies 10x that figure in quantity-matched items alone. The true total including catalog-only distributors is larger still.

---

### What the system cannot do yet

- **Fisher and VWR have no prior-year quantities.** We know prices differ; we cannot estimate how much a researcher would save in dollars without usage data.
- **Pack-size normalization is not implemented.** A &#36;43 item in a case of 200 and a &#36;43 item individually wrapped are not the same price. The LLM reasoning prompt flags this, but the retrieval layer does not filter on it.
- **No feedback loop.** If the LLM picks the wrong equivalent product, there is no mechanism to flag and learn from the error.
- **No UI yet.** Currently a Jupyter cell. FastAPI + HTML wrapper is the next step for researcher access.

---

### Stakeholder questions and answers

| Question | Answer |
|---|---|
| Can we build a RAG? | Yes. Working end-to-end above. |
| Small normalizing model or big reasoning model? | qwen2.5:7b wins the benchmark. It normalizes AND reasons. No separate normalizer needed. |
| How fast is each query? | Exact match: <50ms. Semantic + LLM: 2-12s total. Acceptable for a web tool. |

| Is it scalable? | FAISS handles 6M vectors in ~1GB of RAM. LLM cost is &#36;0 on institutional NIM. |
| What happens when catalogs update? | Re-embed once per catalog cycle (~35 min on GPU). Incremental adds are near-instant. |
| Can we trust the AI's match decisions? | For the reasoning task: yes, qwen2.5:7b passed 100% of equivalence tests. For retrieval: recall depends on description quality (Fisher/VWR terse descriptions are a known gap). |

---

### Recommended next steps

1. **Build the FastAPI wrapper** around `ask_procurement()`. Minimal HTML form - product name input, results table output. This is the researcher-facing tool.
2. **Pack-size normalization.** Before passing candidates to the LLM, normalize price to per-unit using the `qsu` and `uom` fields. This removes a whole class of false equivalences.
3. **Evaluation set.** 20-30 hand-curated queries with known answers. Used to measure regression when prompts or retrieval change.
4. **Deploy to Azure** once local prototype is validated: Azure App Service + Blob Storage for the FAISS index.
5. **Close the Fisher gap.** 1.8M Fisher rows have no part numbers. Options: (a) accept limited recall; (b) enrich high-value Fisher items with catalog lookup; (c) use UNSPSC category + description similarity as a soft join.

````python
# ============================================================
# Stage 18: Unit Price Normalization
# ============================================================
# Uses the qsu (quantity sold as) and uom columns that exist
# in the Jaggaer export to compute true per-unit prices.
# These columns have 100% coverage across the 6M row catalog.
#
# Result: $161,564 defensible savings (vs $317,583 raw)
#         49.1% of the original figure was pack-size inflation
# ============================================================

import numpy as np

print("=== Stage 18: Unit Price Normalization ===")
print()

# ── Step 1: Convert qsu to numeric ────────────────────────
df_all['qsu_num'] = pd.to_numeric(df_all['qsu'], errors='coerce')

has_qsu = df_all['qsu_num'].notna() & (df_all['qsu_num'] > 0)
print(f"Rows with valid qsu:       {has_qsu.sum():,} of {len(df_all):,} ({100*has_qsu.mean():.1f}%)")

normalized_mask = has_qsu & (df_all['qsu_num'] > 1)
print(f"Rows normalized (qsu > 1): {normalized_mask.sum():,} ({100*normalized_mask.mean():.1f}%)")
print(f"Rows single unit (qsu = 1):{(has_qsu & (df_all['qsu_num']==1)).sum():,}")
print()

# ── Step 2: Compute unit price ─────────────────────────────
df_all['unit_price'] = np.where(
    normalized_mask,
    df_all['price_num'] / df_all['qsu_num'],
    df_all['price_num']
)

print("Unit price sample (multi-unit items):")
sample = df_all[normalized_mask][
    ['distributor','description','price_num','qsu_num','uom','unit_price']
].head(5)
print(sample[['distributor','price_num','qsu_num','uom','unit_price']].to_string())
print()

# ── Step 3: Rebuild cross-distributor match table ──────────
print("Rebuilding cross-distributor match table with unit prices...")

keyed_u = df_all[
    df_all['manufacturer'].notna() &
    df_all['mfr_part_num'].notna() &
    (df_all['unit_price'] > 0) &
    (df_all['unit_price'] < 10_000_000)
].copy()
keyed_u['mfr_key']  = keyed_u['manufacturer'].apply(norm_key)
keyed_u['part_key'] = keyed_u['mfr_part_num'].apply(norm_key)
keyed_u = keyed_u[keyed_u['mfr_key'].notna() & keyed_u['part_key'].notna()]

exact_u = (
    keyed_u
    .groupby(['mfr_key','part_key','distributor'], as_index=False)
    .agg(
        unit_price      =('unit_price',       'min'),
        raw_price       =('price_num',         'min'),
        qsu             =('qsu_num',           'first'),
        uom             =('uom',               'first'),
        description     =('description',       'first'),
        manufacturer    =('manufacturer',      'first'),
        dist_catalog_num=('dist_catalog_num',  'first'),
        qty_sold        =('qty_num',           'max'),
    )
)

span = exact_u.groupby(['mfr_key','part_key'])['distributor'].nunique()
multi = span[span >= 2].index
m = exact_u.set_index(['mfr_key','part_key']).loc[multi].reset_index()

agg_u = (
    m.groupby(['mfr_key','part_key'])
     .agg(
         n_dist        =('distributor',   'nunique'),
         min_unit      =('unit_price',    'min'),
         max_unit      =('unit_price',    'max'),
         min_raw       =('raw_price',     'min'),
         max_raw       =('raw_price',     'max'),
         max_qty_sold  =('qty_sold',      'max'),
         desc          =('description',   'first'),
         manufacturer  =('manufacturer',  'first'),
         uom           =('uom',           'first'),
     )
     .reset_index()
)

agg_u['unit_spread']     = agg_u['max_unit'] - agg_u['min_unit']
agg_u['unit_spread_pct'] = (100 * agg_u['unit_spread'] / agg_u['max_unit']).round(2)
agg_u['raw_spread_pct']  = (100 * (agg_u['max_raw'] - agg_u['min_raw']) / agg_u['max_raw']).round(2)
agg_u['est_savings_unit']= agg_u['unit_spread'] * agg_u['max_qty_sold'].fillna(0)

# ── Step 4: Compare raw vs normalized ─────────────────────
spread_data        = agg_u[agg_u['unit_spread_pct'] > 0]['unit_spread_pct']
total_unit_savings = float(agg_u['est_savings_unit'].sum())
total_raw_savings  = float(((agg_u['max_raw'] - agg_u['min_raw']) * agg_u['max_qty_sold'].fillna(0)).sum())
reduction          = total_raw_savings - total_unit_savings

print(f"Cross-distributor matched pairs:   {len(agg_u):,}")
print()
print("Unit-price spread distribution:")
print(spread_data.describe(percentiles=[.25,.5,.75,.9,.99]).round(2).to_string())
print()
print(f"Total savings - RAW prices:        ${total_raw_savings:,.0f}")
print(f"Total savings - UNIT prices:       ${total_unit_savings:,.0f}")
print(f"Pack-size inflation removed:       ${reduction:,.0f} ({100*reduction/total_raw_savings:.1f}%)")
print()

# ── Step 5: Top 25 by unit-price savings ──────────────────
top25_u = agg_u[agg_u['est_savings_unit'] > 0].nlargest(25, 'est_savings_unit')
print("TOP 25 BY UNIT-PRICE SAVINGS:")
print(f"{'Rank':<5} {'Est Savings':>12} {'Spread%':>8}  Description")
print("-" * 90)
for i, (_, r) in enumerate(top25_u.iterrows(), 1):
    print(f"{i:<5} ${r['est_savings_unit']:>11,.0f} {r['unit_spread_pct']:>7.1f}%  {str(r['desc'])[:58]}")

print()
print("=" * 60)
print("VERDICT: Defensible unit-normalized savings: ${:,.0f}".format(total_unit_savings))
print("This is the number to show leadership.")
print("Original raw figure was ${:,.0f} — {:.1f}% was pack-size inflation.".format(
    total_raw_savings, 100*reduction/total_raw_savings))
print("=" * 60)

# Make agg_u available for Stage 19
agg_unit_normalized = agg_u.copy()
print()
print("agg_unit_normalized is now available for Stage 19 (dashboard export).")
````
